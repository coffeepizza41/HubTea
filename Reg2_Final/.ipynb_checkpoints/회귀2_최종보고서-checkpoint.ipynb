# 3. 데이터 전처리

## (1) Encoding

### 데이터 변환

```python
df['Gender'].replace(to_replace={'Male':0,'Female':1},
             inplace=True)
df['Vehicle_Damage'].replace(to_replace={'No':0,'Yes':1},
             inplace=True)
df['Vehicle_Age'].replace(to_replace={'< 1 Year':0,'1-2 Year':1,'> 2 Years':2},
             inplace=True)
```

## (3) 변수 준비하기

### 반응변수와 설명변수로 분리 

```python
df=df.drop(columns=['id']) 
```

```python
y=df.Response                   
x=df.drop(columns=['Response']) 
```

### 불균형 데이터 처리

```python
randomsample= RandomOverSampler()
x_new,y_new=randomsample.fit_sample(x,y)

print('Original dataset shape {}'.format(Counter(y)))
print('Resampled dataset shape {}'.format(Counter(y_new)))
sns.countplot(y_new)
```

## (4) 데이터 나누기 - train/test data

```python
xtrain,xtest,ytrain,ytest=train_test_split(x_new,y_new,test_size=.3,random_state=42)
```

## (5) 데이터 스케일링

```python
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

# 4. 로지스틱 모형 적합

## (1) LogisticRegression() 이용

### 모형적합

```python
lr=LogisticRegression()
lr=lr.fit(xtrain, ytrain)
print("절편:",lr.intercept_) 
print("회귀계수:", lr.coef_) 
```

### 예측

```python
pred=lr.predict(xtest)
print("Logistic regression Reports:\n", classification_report(ytest, pred))
```

### 정확도(Accuracy)

```python
acc=accuracy_score(ytest,pred)
print("Accuracy of Logistic Regression" + ' : ' + str(acc)) 
```

### ROC_AUC

```python
AUC=roc_auc_score(pred,ytest)
print("ROC_AUC Score of Logistic Regression:" + ' : ' + str(AUC))
```

### ROC_Curve

```python
lr_probability = lr.predict_proba(xtest)[:,1]
fpr, tpr, _ = roc_curve(ytest, lr_probability)

plt.title('Logistic Regression ROC curve')
plt.xlabel('FPR (Precision)')
plt.ylabel('TPR (Recall)')

plt.plot(fpr,tpr)
plt.plot((0,1), ls='dashed',color='red')
plt.show()
```

### Confusion matrix

```python
cm=confusion_matrix(ytest,pred)
print(cm)
sns.heatmap(cm,annot=True)
```

## (2) sm의 Logit 이용

### 모형적합

```python
logit = sm.logit(ytrain, xtrain)
result = logit.fit(method="newton")
```

### 모형적합 결과

```python
result.summary2()
```

### cf. GLM

```python
logit = sm.GLM(ytrain, xtrain, family=sm.families.Binomial())
result = logit.fit(method="newton")
```

```python
result.summary2()
```

### 오즈비

```python
np.exp(result.params)
```
